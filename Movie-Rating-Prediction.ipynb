```python

# Example DataFrame
data = pd.DataFrame({
    'Year': ['2012', '2019', '2014', '2016', '2018'],
    'Rating': [8.5, 7.2, 9.1, 6.5, 8.0]
})

# Check columns
print(data.columns)

# Group by 'Year' and calculate mean of 'Rating'
data1 = data.groupby('Year')['Rating'].mean()\
    .sort_values(ascending=False)\
    .reset_index()\
    .set_index('Year')

print(data1)
```

```python
# Sample DataFrame
data = pd.DataFrame({
    'Year': ['2012', '2019', '2014', '2016', '2018'],
    'Rating': [8.5, 7.2, 9.1, 6.5, 8.0]
})

# Group by 'Year' and calculate mean of 'Rating'
data1 = data.groupby('Year')['Rating'].mean()\
    .sort_values(ascending=False)\
    .reset_index()\
    .set_index('Year')

plt.figure(figsize=(10, 5))

# Create a barplot
sns.barplot(x=data1.index, y=data1['Rating'])

plt.title("Average Rating by Year")
plt.xlabel("Year")
plt.ylabel("Average Rating")
plt.show()
```
